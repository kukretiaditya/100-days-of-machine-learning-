```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('orders.csv')
time = pd.read_csv('messages.csv')
```

```python
data.head(), time.head()
```

```python
data.info(), time.info()
```

```python
#converting to datetime datatye
data['date']= pd.to_datetime(data['date'])
```

```python
data.info()
```

```python
#extract year from date
data.head()
data['data_year_extract'] = data['date'].dt.year
data.head()
data.sample(5)
```

```python
#extract month
data['data_month_extract'] = data['date'].dt.month
data.head()
```

```python
#extract day
data['data_day_extract'] = data['date'].dt.day
data.head()
```

```python
#day of week
data['date_dow'] = data['date'].dt.dayofweek
data.head()
```

```python
#day of week
data['date_day name '] = data['date'].dt.day_name()
data.head()
```

```python
#week of the year
data['date_week'] = data['date'].dt.isocalendar().week
data.head()
```

```python
#extract quarter
data['quarter'] = data['date'].dt.quarter
data['quarter']
```

```python
#extract semester
data['semester'] = np.where(data['quarter'].isin([1,2]),1,2)
data['semester']
```

```python
#extract time between dates
import datetime
today = datetime.datetime.today()
today
```

```python
today - data['date']
```

```python
(today - data['date']).dt.days
```

```python
# passed
np.round((today - data['date']) / np.timedelta64(1, 'D') ,0)
```

```python
time['hour'] = time['date'].dt.hour
time['min'] = time['date'].dt.minute
time['sec'] = time['date'].dt.second
time.head()
```

```python
#time diff
today - time['date']
```

```python
#in seconds
(today - time['date'])/np.timedelta64(1,'s')
```

```python
#in min
(today - time['date'])/np.timedelta64(1,'m')
```

```python
#in hours
(today - time['date'])/np.timedelta64(1,'h')
```
